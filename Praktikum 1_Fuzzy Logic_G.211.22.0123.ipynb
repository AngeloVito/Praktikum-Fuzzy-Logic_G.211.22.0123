```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt
```

**Segitiga**

```python
x = np.arange(11)
x
```

```python
# Triangular () com a=0, b=5, c=10.
mfx = fuzzy.trimf(x, [0, 5, 10])
mfx
```

```python
plt.plot(x, mfx, 'r', linewidth=1.5, label='Triangular')
plt.fill(x, mfx, 'b', alpha=0.5)
plt.legend()
plt.show()
```

**Trapesium**

```python
# Trapezoidal - trapmf com a=0, b=4, c=7, d=10
mft = fuzzy.trapmf(x,[0,4,7,10])
mft
```

```python
plt.plot(x, mft, 'g', linewidth=1.5, label='Trapezoidal')
plt.fill(x, mft, 'y', alpha=0.5)
plt.legend()
plt.show()
```

**Gaussian**

```python
val = np.arange(-10, 11)
```

```python
# Gaussian - gaussmf
mfg = fuzzy.gaussmf(val, 0.2, 5)
mfg
```

```python
plt.plot(val, mfg, 'b', linewidth=1.5, label='Gaussian')
plt.fill(val, mfg, 'b', alpha=0.5)
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 500)
```

```python
mean1, sigma1 = 3, 1.0
mean2, sigma2 = 7, 0.8
mean3, sigma3 = 5, 1.5
```

```python
membership1 = fuzzy.gaussmf(x, mean1, sigma1)
membership2 = fuzzy.gaussmf(x, mean2, sigma2)
membership3 = fuzzy.gaussmf(x, mean3, sigma3)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, membership1, label=f'Fuzzy Set 1 (mean={mean1}, sigma={sigma1})',color='blue')
plt.plot(x, membership2, label=f'Fuzzy Set 2 (mean={mean2}, sigma={sigma2})',color='green')
plt.plot(x, membership3, label=f'Fuzzy Set 3 (mean={mean3}, sigma={sigma3})',color='red')

plt.title('Gaussian Fuzzy Fungsi Keanggotaan')
plt.xlabel('Input (x)')
plt.ylabel('Derajat Keanggotaan')
plt.grid(True)
plt.legend()
plt.ylim(0, 1.1) # Ensure y-axis shows full membership range
plt.show()
```

**Sigmoid**

```python
x = np.arange(0, 11, 0.1)
```

```python
b_positive = 5   # Titik Pusat yang Derajat Keanggotaannya 0.5
c_positive = 1   # Kelengkungan Kurva
sigmf_positive = fuzzy.sigmf(x, b_positive, c_positive)
```

```python
b_negative = 5
c_negative = -1
sigmf_negative = fuzzy.sigmf(x, b_negative, c_negative)
```

```python
# Plotting the Membership Functions
plt.figure(figsize=(10, 6))

plt.plot(x, sigmf_positive, 'b', linewidth=1.5, label='Positive Sigmoid (c > 0)')
plt.plot(x, sigmf_negative, 'r', linewidth=1.5, label='Negative Sigmoid (c < 0)')

plt.title('Fuzzy Sigmoid Fungsi Keanggotaan')
plt.xlabel('Input (x)')
plt.ylabel('Berlangganan')
plt.grid(True)
plt.legend()
plt.show()
```
