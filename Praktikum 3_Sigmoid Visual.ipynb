```python
x_usia = np.arange(18, 58, 1)
x_suhubadan = np.arange(27.5, 45.0, 1)
x_kondnapas = np.arange(14, 24, 1)
x_jmlantbd = np.arange(0, 10, 1)
x_status = np.arange(0, 100, 1)

usia_muda = fuzz.trapmf(x_usia, [18, 18, 30, 35])
usia_pibaya = fuzz.trapmf(x_usia, [30, 36, 45, 50])
usia_tua = fuzz.trapmf(x_usia, [45, 50, 58, 58])

suhu_hipo = fuzz.trapmf(x_suhubadan, [27.5, 27.5, 34.5, 35])
suhu_normal = fuzz.trapmf(x_suhubadan, [35, 35.5, 36.5, 37])
suhu_hiper = fuzz.trapmf(x_suhubadan, [36.5, 37.5, 45.0, 45.0])

kondnapas_normal = fuzz.trapmf(x_kondnapas, [14, 14, 16, 17])
kondnapas_moderat = fuzz.trapmf(x_kondnapas, [16, 18, 19, 20])
kondnapas_abnormal = fuzz.trapmf(x_kondnapas, [19, 20, 24, 24])

jmlantbd_normal = fuzz.trapmf(x_jmlantbd, [0, 0.1, 1, 1.5])
jmlantbd_moderat = fuzz.trapmf(x_jmlantbd, [1.5, 2, 3, 4])
jmlantbd_abnormal = fuzz.trapmf(x_jmlantbd, [3.5, 4, 8, 8.0])

status_negatif = fuzz.trapmf(x_status, [0, 0, 49, 50])
status_positif = fuzz.trapmf(x_status, [50, 51, 100, 100])
```

```python
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, figsize=(8, 10))

ax0.plot(x_usia, usia_muda, 'b', linewidth=1.5, label='Muda')
ax0.plot(x_usia, usia_pibaya, 'r', linewidth=1.5, label='Paruhbaya')
ax0.plot(x_usia, usia_tua, 'g', linewidth=1.5, label='Tua')
ax0.set_title('Indeks Usia Pasien')
ax0.legend()

ax1.plot(x_suhubadan, suhu_hipo, 'b', linewidth=1.5, label='Hipotermia')
ax1.plot(x_suhubadan, suhu_normal, 'r', linewidth=1.5, label='Normal')
ax1.plot(x_suhubadan, suhu_hiper, 'g', linewidth=1.5, label='Hipertemia')
ax1.set_title('Indeks Suhu Badan')
ax1.legend()

ax2.plot(x_kondnapas, kondnapas_normal, 'b', linewidth=1.5, label='Normal')
ax2.plot(x_kondnapas, kondnapas_moderat, 'r', linewidth=1.5, label='Moderat')
ax2.plot(x_kondnapas, kondnapas_abnormal, 'g', linewidth=1.5, label='Abnormal')
ax2.set_title('Indeks Kondisi Pernapasan')
ax2.legend()

ax3.plot(x_jmlantbd, jmlantbd_normal, 'b', linewidth=1.5, label='Normal')
ax3.plot(x_jmlantbd, jmlantbd_moderat, 'r', linewidth=1.5, label='Moderat')
ax3.plot(x_jmlantbd, jmlantbd_abnormal, 'g', linewidth=1.5, label='Abnormal')
ax3.set_title('Indeks Jumlah Antibodi')
ax3.legend()

# Plot variabel status pasien
ax4.plot(x_status, status_negatif, 'b', linewidth=1.5, label='Negatif')
ax4.plot(x_status, status_positif, 'r', linewidth=1.5, label='Positif')
ax4.set_title('Indeks Status Pasien')
ax4.legend()

for ax in (ax0, ax1, ax2, ax3, ax4):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# --- TEKANAN DARAH (mmHg) ---
x_bp = np.arange(80, 181, 1)
bp_low = fuzz.zmf(x_bp, 100, 120)
bp_normal = fuzz.gaussmf(x_bp, 120, 10)
bp_high = fuzz.smf(x_bp, 130, 150)

# --- DENYUT JANTUNG (bpm) ---
x_hr = np.arange(40, 141, 1)
hr_low = fuzz.zmf(x_hr, 60, 80)
hr_normal = fuzz.gaussmf(x_hr, 85, 10)
hr_high = fuzz.smf(x_hr, 90, 110)

# --- SUHU TUBUH (°C) ---
x_temp = np.arange(35, 42, 0.1)
temp_low = fuzz.zmf(x_temp, 36, 36.5)
temp_normal = fuzz.trimf(x_temp, [36.5, 37, 37.5])
temp_high = fuzz.smf(x_temp, 37.5, 38)

# --- GULA DARAH (mg/dL) ---
x_glucose = np.arange(80, 301, 1)
glucose_low = fuzz.zmf(x_glucose, 100, 120)
glucose_normal = fuzz.gaussmf(x_glucose, 130, 20)
glucose_high = fuzz.smf(x_glucose, 150, 200)

# --- RISIKO PENYAKIT (OUTPUT) ---
x_risk = np.arange(0, 101, 1)
risk_low = fuzz.zmf(x_risk, 30, 50)
risk_medium = fuzz.gaussmf(x_risk, 50, 15)
risk_high = fuzz.smf(x_risk, 60, 80)

# --- VISUALISASI ---
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
axs = axs.flatten()

# 1. Tekanan Darah
axs[0].plot(x_bp, bp_low, 'b', label='Low')
axs[0].plot(x_bp, bp_normal, 'g', label='Normal')
axs[0].plot(x_bp, bp_high, 'r', label='High')
axs[0].set_title('Tekanan Darah (mmHg)')
axs[0].legend()

# 2. Denyut Jantung
axs[1].plot(x_hr, hr_low, 'b', label='Low')
axs[1].plot(x_hr, hr_normal, 'g', label='Normal')
axs[1].plot(x_hr, hr_high, 'r', label='High')
axs[1].set_title('Denyut Jantung (bpm)')
axs[1].legend()

# 3. Suhu Tubuh
axs[2].plot(x_temp, temp_low, 'b', label='Low')
axs[2].plot(x_temp, temp_normal, 'g', label='Normal')
axs[2].plot(x_temp, temp_high, 'r', label='High')
axs[2].set_title('Suhu Tubuh (°C)')
axs[2].legend()

# 4. Gula Darah
axs[3].plot(x_glucose, glucose_low, 'b', label='Low')
axs[3].plot(x_glucose, glucose_normal, 'g', label='Normal')
axs[3].plot(x_glucose, glucose_high, 'r', label='High')
axs[3].set_title('Gula Darah (mg/dL)')
axs[3].legend()

# 5. Risiko Penyakit
axs[4].plot(x_risk, risk_low, 'b', label='Low Risk')
axs[4].plot(x_risk, risk_medium, 'g', label='Medium Risk')
axs[4].plot(x_risk, risk_high, 'r', label='High Risk')
axs[4].set_title('Risiko Penyakit (Output)')
axs[4].legend()

# Hilangkan subplot kosong terakhir
fig.delaxes(axs[5])

plt.tight_layout()
plt.show()
```

```python
# Menentukan Domain Suhu
x_suhu = np.arange(0, 101, 1)   # Domain suhu mesin 0 - 100 C
```

```python
# Kriteria Suhu
dingin = fuzz.trimf(x_suhu, [0, 0, 40])   # segitiga (0 - 40)
normal = fuzz.trapmf(x_suhu, [30, 45, 65, 80])   # trapesium (30 - 80)
panas = fuzz.trimf(x_suhu, [70, 100, 100])   # segitiga (70 - 100)
```

```python
# Visualisasi
plt.figure(figsize=(8,5))
plt.plot(x_suhu, dingin, 'b-', label='Dingin')
plt.plot(x_suhu, normal, 'g-', label='Normal')
plt.plot(x_suhu, panas, 'r-', label='Panas')
plt.title('Derajat Keanggotaan Fuzzy Suhu Mesin Industri (scikit-fuzzy)')
plt.xlabel('Suhu (°C)')
plt.ylabel('Derajat Keanggotaan (μ)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
suhu_input = 38

# Hitung derajat keanggotaan
μ_dingin = fuzz.interp_membership(x_suhu, dingin, suhu_input)
μ_normal = fuzz.interp_membership(x_suhu, normal, suhu_input)
μ_panas = fuzz.interp_membership(x_suhu, panas, suhu_input)

# Tampilkan hasil
print(f"Suhu mesin: {suhu_input}°C")
print(f"μ Dingin = {μ_dingin:.2f}")
print(f"μ Normal = {μ_normal:.2f}")
print(f"μ Panas = {μ_panas:.2f}")
```
